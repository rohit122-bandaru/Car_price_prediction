```python
import numpy as np
import pandas as pd
import sklearn
```

```python
# Importing LinearRegression model
from sklearn.linear_model import LinearRegression
```

```python
# Importing accuracy score function
from sklearn import metrics
```

```python
# Importing train test split function
from sklearn.model_selection import train_test_split
```

```python
# Setting up the data
# Loading the data set
car_dataset = pd.read_csv("car data.csv")
```

```python
# Displaying the first 5 rows
car_dataset.head()
```

```python
car_dataset.info()
```

```python
# Checking the null values
car_dataset.isnull().sum()
```

```python
# Checking the distribution of categorical data
pd.unique(car_dataset["Fuel_Type"])
car_dataset["Fuel_Type"].value_counts().reset_index()
```

```python
# Checking the distribution of categorical data
pd.unique(car_dataset["Seller_Type"])
car_dataset["Seller_Type"].value_counts().reset_index()
```

```python
# Checking the distribution of categorical data
pd.unique(car_dataset["Transmission"])
car_dataset["Transmission"].value_counts().reset_index()
```

```python
# Encoding categorical columns
car_dataset.replace({'Fuel_Type': {'Petrol': 0, 'Diesel': 1, 'CNG': 2}}, inplace=True)
car_dataset.replace({'Seller_Type': {'Dealer': 0, 'Individual': 1}}, inplace=True)
car_dataset.replace({'Transmission': {'Manual': 0, 'Automatic': 1}}, inplace=True)
```

```python
car_dataset
```

```python
# Creating Model
X = car_dataset.drop(['Car_Name','Selling_Price'] , axis = 1)
Y = car_dataset['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Print shapes to verify
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)

# Model Training
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Predict of training data
training_data_prediction = model.predict(X_train)
training_data_accuracy = metrics.r2_score(Y_train ,training_data_prediction)
print(training_data_accuracy)
```

```python
# Predict of testing data
testing_data_prediction = model.predict(X_test)
testing_data_accuracy = metrics.r2_score(Y_test ,testing_data_prediction)
print(testing_data_accuracy)
```

```python
# predictive system
input_data = (2014,5.59,27000,0,0,0,0)
input_data_as_numpy_array = np.array(input_data)
# Reshape the numpy array
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)
```

```python
prediction = model.predict(input_data_reshape )
print(prediction)
```
